### Using LabelEncoder to encode categorical string data i.e. 'gender' and 'country' features

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,add_dummy_feature
```

```python
label_encoder_gender = LabelEncoder()
label_encoder_country = LabelEncoder()
```

```python
df['gender'] = label_encoder_gender.fit_transform(df['gender'])
df['country'] = label_encoder_country.fit_transform(df['country'])
```

```python
df
```

```python
df.test = df.test + 5
```

```python
df
```

```python
countries = ['Gabon','Ghana','Nigeria']
```

```python
df['country_name'] = df.country.apply(lambda x: countries[x])
```

```python
df.drop('country_name',axis=1,inplace=True)
```

```python
df
```

```python
age_mean = df.age.mean()
age_dev = df.age.std()
df['age_std'] = (df.age - age_mean) / age_dev #Scaled using Standard Scaling (standardization)
```

```python
min_age = df.age.min()
max_age = df.age.max()
range_of_age = max_age - min_age
df['age_norm'] = (df.age - min_age) / range_of_age #Scaled using Standard Scaling (standardization)
```
